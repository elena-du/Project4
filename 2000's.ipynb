```python
def get_corpous(artist_input):
    artist_name = artist_input
    api = genius.Genius(geniusCreds)
    artist = api.search_artist(artist_name, max_songs=60)
    artist.save_lyrics()
    with open('Lyrics_'+artist_input+'.json') as json_data:
        data = json.load(json_data)
    artist_corpous = [] 
    for i in range(0,10):
        artist_corpous.append((data['songs'][i]['lyrics']))   
    return artist_corpous
```

```python
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
import re
import string
def cleaning_data(artist_names):
    cleaned = []
    for artist in artist_names:
        word_tokenize(artist)
        alphanumeric = lambda artist: re.sub('\w*\d\w*', ' ', artist)
        clean_text=re.sub('[%s]' %re.escape(string.punctuation),'',artist)
        clean_text = clean_text.lower()
        clean_text =re.sub('\w*\d\w*','',clean_text)
        ##clean_text = word_tokenize(clean_text)
        cleaned.append(clean_text)
    return cleaned
```

```python
def nmf(artist_cleaned):
    vectorizer = CountVectorizer(stop_words = 'english')
    doc_word = vectorizer.fit_transform(artist_cleaned)
    pd.DataFrame(doc_word.toarray(), columns=vectorizer.get_feature_names()).head(10)
    nmf_model = NMF(8)
    doc_topic = nmf_model.fit_transform(doc_word)
    topic_word = pd.DataFrame(nmf_model.components_.round(3),
             index = ["component_1","component_2", "component_3", "component_4","component_5","component_6","component_7","component_8"],
             columns = vectorizer.get_feature_names())
    topic_word
    def display_topics(model, feature_names, no_top_words, topic_names=None):
        for ix, topic in enumerate(model.components_):
            if not topic_names or not topic_names[ix]:
                print("\nTopic ", ix)
            else:
                print("\nTopic: '",topic_names[ix],"'")
            print(", ".join([feature_names[i]
                            for i in topic.argsort()[:-no_top_words - 1:-1]]))
    display_topics(nmf_model, vectorizer.get_feature_names(), 8)
    Vt = pd.DataFrame(doc_topic.round(8),
             index = artist_cleaned,
             columns = ["component_1","component_2", "component_3", "component_4","component_5","component_6","component_7","component_8" ])
    Vt
    return  Vt
```
